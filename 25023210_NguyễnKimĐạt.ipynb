```python
def nhap():
    n = int(input("Nhập một số n (n <= 1000): "))
    while n > 1000:
        print("Vui lòng nhập số nhỏ hơn hoặc bằng 1000")
        n = int(input("Nhập lại n: "))
    return n

n = nhap()

ans = 0
for i in range(1, n + 1):
    ans += i

print(f"Tổng các số hạng từ 1 đến n là: {ans}")
```

```python
def nguyenduong():
    n = int(input("Nhập 1 số nguyên dương: "))
    while n <= 0:
        print("Vui lòng nhập số dương")
        n = int(input("Nhập số n: "))
    return n
n = nguyenduong()
if n == 1:
    print("1 không phải số nguyên tố")
else:
    for i in range(2, int(n**(1/2))):
        if n % i == 0:
            print(f"{n} không phải số nguyên tố")
            break
    else:
        print(f"{n} là số nguyên tố")
```

```python
from math import *

def ngto(n):
    
    if( n < 2): return False;

    _n = int(sqrt(n))

    for i in range( 2 , _n + 1):
        if n % i == 0:
            return False
        

    return True

def nhap():
    
    sum = 0;

    a, b = map(int, input().split()) 
    while a >= b or b <0:
        print('nhập lại dữ liệu!')
        a, b = map(int, input().split())

    for i in range( a , b + 1):
        if (ngto(i)): sum += i;

    return sum;

while True:
    print(nhap());
```

```python
n = int(input())

def collatz(v):
    dem = 0
    while v != 1:
        if v % 2 == 0:
            v = v / 2
        else:
            v = 3*v +1
        dem += 1
    return dem
so_i = 0
max_collatz = 0
for i in range(1, n+1):
    if collatz(i) > max_collatz:
        so_i = i
        max_collatz = collatz(i)
print(so_i, max_collatz)

```

```python
def uoc(n):

    while n < 0:
        return False
    lst = []
    for i in range(1, n):
        if n % i == 0:
            lst.append(i)
    return sum(lst)
def nhap():
    a, b = map(int, input().split())
    c = uoc(a)
    d = uoc(b)
    if c == b and d == a:
        print("true")
    else:
        print("false")
nhap()
```
